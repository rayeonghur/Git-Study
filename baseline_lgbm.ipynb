#Train 및 Validation Set 분할

```python
#학습 데이터셋, 검증 데이터셋으로 분할.

#1. 두 set을 구분하기 위한 난수 생성.
sample = list(range(len(finalDf))) #데이터셋 길이에 맞는 정수 리스트 생성.
random.shuffle(sample) #해당 리스트 마구잡이로 섞음.

#2. train 및 vaildation set 50:50으로 분할.
trainDf = finalDf.iloc[sample,].head(len(finalDf)//2) #head. 섞인 데이터의 상위50%를 train으로 지정.
valDf = finalDf.iloc[sample,].tail(len(finalDf)//2) #tail. 섞인 데이터의 하위50%를 validation으로 지정.
```

#하이퍼파라미터 튜닝 : 그리드 서치 모델 생성

LGBM의 하이퍼파라미터

- learning_rate : 부스팅 각 이터레이션마다 곱해지는 가중치.<br>
    - 보통 작을수록 모형성능 향상. 학습시간 길어짐.
    - 하이퍼파라미터 튜닝시에는 0.1~0.3정도의 값, 최종 모형 학습시에는 0.05이하의 값을 사용하는 것이 좋음.

- max_depth : 각 트리의 최대 깊이.
    - 클수록 학습데이터에 피팅이 잘 되지만, 오버피팅 가능성도 높아짐.
    - 보통 3~12 사용
    - 가장 먼저 튜닝해야 할 하이퍼파라미터

- boosting : 부스팅 방법 결정.
    - 디폴트 : gbdt.

- objective
- metric
- num_leaves
- feature_fraction : 각 이터레이션에 사용되는 column의 비율.
    - 보통 0.7 정도의 값.
    - 세세히 튜닝할 필요는 없음.
    - 모형 성능, 학습시간에 영향

- bagging_fraction : 각 이터레이션에 사용되는 행의 비율.
    - feature_fraction과 마찬가지.

- 바꿀 파라미터 정의할 때의 각 파라미터 명

  - param = {
    'learning_rate' : 0.03,
    'max_depth' : 2,
    'boosting' : 'gbdt',
    'objective' : 'binary',
    'metric' : 'auc',
    'num_leaves' : 5,
    'feature_fraction' : 0.5,
    'bagging_fraction' : 0.5,
    'num_class' : 3
}

- 고정 파라미터 정의할 떄의 각 파라미터 명
  - XGBClassifier(n_estimators=1000, random_state=156, learning_rate=0.02, max_depth=5, min_child_weight=1, colsample_bytree=0.75, reg_alpha=0.03)

```python
from lightgbm import LGBMClassifier
#주요 하이퍼 파라미터 설정! => 결과를 바꿔가면서 제일 좋은 걸로 확인하고 싶은 파라미터 정의

#연습삼아 우선 대강 세팅.


params = {
    'max_depth' : [2,4,7],
    'num_leaves' : [5,6]
}

num_round = 10 #학습을 몇번 수행할지 세팅! 10 -> 10번 학습 진행!

#분류기 생성
#안 바꿀 파라미터들 입력
#https://velog.io/@wltn39/XGBoost-LightGBM-%EB%B6%84%EB%A5%98%ED%95%99%EC%8A%B5
LGBM_clf = LGBMClassifier(n_estimators=10,boosting_type='gbdt',num_leaves=5,learning_rate=0.03)

#GridSearch
#estimator : 분류기 지정, param_grid : 측정할 파라미터 지정, cv : cross validation (교차 검증)
gridSearch = GridSearchCV(estimator = LGBM_clf, param_grid=params,cv = 5)
gridSearch.fit(trainDf.iloc[:,:-1],trainDf.iloc[:,-1])
```

```python
#최적의 성능을 가진 모델
print(gridSearch.best_estimator_)
#최적일 경우의 값
print(gridSearch.best_score_)
#최적일 경우의 prameters
print(gridSearch.best_params_)
```

#그냥 모델 생성

```python
#최적의 성능을 가진 모델 : gridSearch.best_estimator_
```

```python
#그냥 학습 수행해 모델 생성!
param = {
    'learning_rate' : 0.03,
    'max_depth' : 3,
    'boosting' : 'gbdt',
    'objective' : 'binary',
    'metric' : 'auc',
    'num_leaves' : 5,
    'feature_fraction' : 0.5,
    'bagging_fraction' : 0.5,
    'num_class' : 1
}
lgbmbst = lgb.train(params=param, train_set=dtrain, num_boost_round=num_round, valid_sets=dtest)
```
