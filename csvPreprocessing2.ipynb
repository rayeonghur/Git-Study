#ATT&CK Tactic 처리

```python
#ATT&CK Tactic 컬럼에 총 아이템 수 (중복 제외)
attTactic=[]
for item in csvPre['ATT&CK Tactic']:
  if isinstance(item, str):
     #세미콜론으로 아이템 추출
     split_values = item.split(';')
     attTactic.extend(split_values)
#중복 제거
attTactic = list(set([value.strip() for value in attTactic]))
print(attTactic)
print("attTactic total item number: ", len(attTactic));
```

```python
#원핫 인코딩
import pandas as pd
one_hot_encoded = csvPre['ATT&CK Tactic'].str.get_dummies(sep=';')
result_df = pd.concat([csvPre, one_hot_encoded.add_prefix("ATT_Tactic_")], axis=1)
result_df.head()
result_df.to_csv(newFileName, index=False)
```

#ATT&CK Technique 처리

```python
#ATT&CK Technique열 총 항목 수
import re
attTechnique=[]

#대괄호 안의 항목을 추출하는 함수
def extract_bracket_content(s):
    return re.findall(r'\[([^\]]+)\]', s)


for item in csvPre["ATT&CK Technique"]:
    if isinstance(item, str):
        parts = item.split(';')
        for part in parts:
            attTechnique.extend(extract_bracket_content(part))

# 중복제거
attTechnique = list(set([value.strip() for value in attTechnique]))
print(attTechnique)
print("attTechnique total item number: ", len(attTechnique));
```

```python
result_df.head()
```

```python
#원핫인코딩
df=existing_df = pd.read_csv(newFileName)
one_hot_encoded = pd.DataFrame(columns=['ATT_Technique_' + technique for technique in attTechnique])

for item in df["ATT&CK Technique"]:
    row_data = {('ATT_Technique_' + technique): 0 for technique in attTechnique}
    if isinstance(item, str):
        parts = item.split(';')
        for part in parts:
            techniques = extract_bracket_content(part)
            for technique in techniques:
                row_data['ATT_Technique_' + technique.strip()] = 1
    one_hot_encoded = pd.concat([one_hot_encoded, pd.DataFrame([row_data])], ignore_index=True)

# 결과를 기존 데이터프레임에 병합
result_df = pd.concat([df, one_hot_encoded], axis=1)
result_df.to_csv(newFileName, index=False)
result_df.head()
```

#Namespace 처리

```python
# 상위 디렉토리 기준으로만 개수
namespace=[]
for item in csvPre["Namespace"]:
    # 값이 문자열인 경우만 처리 (NaN 등의 예외 처리)
    if isinstance(item, str):
        # 문자열을 ;으로 나눔
        parts = item.split(';')
        # 각 부분에서 가장 상위 디렉토리 추출
        top_directories = [part.split('/')[0].strip() for part in parts]
        namespace.extend(top_directories)
namespace = list(set(namespace))
print(namespace)
print("namespace total item number: ", len(namespace));
```

```python
# 전체 디렉토리 기준으로만 개수
namespaceFull=[]
for item in csvPre["Namespace"]:
    # 값이 문자열인 경우만 처리 (NaN 등의 예외 처리)
    if isinstance(item, str):
        # 문자열을 ;으로 나눔
        parts = item.split(';')
        top_directories = [part.strip() for part in parts]
        namespaceFull.extend(top_directories)
namespaceFull = list(set(namespaceFull))
print(namespaceFull)
print("namespace total item number: ", len(namespaceFull));
```

```python
#상위 디렉토리로 원핫인코딩
import pandas as pd

# 상위 디렉토리 추출 함수
def extract_top_directories(s):
    parts = s.split(';')
    return [part.split('/')[0].strip() for part in parts]

# 상위 디렉토리 리스트 생성
namespace = []
for item in df["Namespace"]:
    if isinstance(item, str):
        namespace.extend(extract_top_directories(item))
namespace = list(set(namespace))

# 각 상위 디렉토리별로 개수 세기
for top_dir in namespace:
    # 새로운 열 이름: 'namespace_' + 상위 디렉토리 이름
    column_name = 'namespace_' + top_dir
    # 각 행에 대해 상위 디렉토리의 개수 계산
    df[column_name] = df["Namespace"].apply(lambda s: s.count(top_dir) if isinstance(s, str) else 0)

# 결과 확인
df.head()
result_df.to_csv(newFileName, index=False)
```
